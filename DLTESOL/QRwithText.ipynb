```python
# Install required libraries
# !pip install qrcode[pil] Pillow

# Import required libraries
import qrcode
from PIL import Image, ImageDraw, ImageFont
from IPython.display import display

def create_qr_with_text_center(text, qr_text):
    # Determine the appropriate version based on the length of the QR code data
    version = min(40, max(1, (len(qr_text) // 25) + 1))  # Ensure version stays between 1 and 40

    # Generate QR code
    qr = qrcode.QRCode(
        version=version,
        error_correction=qrcode.constants.ERROR_CORRECT_H,
        box_size=10,
        border=4,
    )
    qr.add_data(qr_text)
    qr.make(fit=True)

    # Create an image from the QR Code instance
    img = qr.make_image(fill='black', back_color='white').convert('RGB')

    # Load a font
    try:
        font = ImageFont.truetype("arial.ttf", 40)  # Use a truetype font
    except IOError:
        font = ImageFont.load_default()  # Fallback to default font

    # Calculate text size and position
    draw = ImageDraw.Draw(img)
    text_width, text_height = draw.textsize(text, font=font)
    img_width, img_height = img.size
    text_x = (img_width - text_width) / 2
    text_y = (img_height - text_height) / 2

    # Clear a rectangle in the center of the QR code to place the text
    rectangle_padding = 10  # Adjust padding around the text as needed
    clear_box = [
        (text_x - rectangle_padding, text_y - rectangle_padding),
        (text_x + text_width + rectangle_padding, text_y + text_height + rectangle_padding)
    ]
    draw.rectangle(clear_box, fill="white")

    # Add text to image
    draw.text((text_x, text_y), text, fill="black", font=font)

    return img

# Example usage
user_text = input("Enter the text to display in the middle of the QR code: ")
qr_text = input("Enter the text/data for the QR code: ")

qr_image = create_qr_with_text_center(user_text, qr_text)
display(qr_image)  # Display the image

# Save the image if needed
qr_image.save("qr_code_with_text.png")
```
