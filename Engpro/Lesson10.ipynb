```python
#@markdown ðŸ”„ Making functions: etts('text'), ktts('text'), btts('text)
%%capture
!pip install gTTS
from gtts import gTTS
from IPython.display import Audio

def etts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "en", tld = "us",
                    slow = False)

  gtts_object.save("E-audio.mp3")
  return Audio("E-audio.mp3")

def ktts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "ko",
                    slow = False)

  gtts_object.save("E-audio.mp3")
  return Audio("E-audio.mp3")

def btts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "en", tld = "co.uk",
                    slow = False)

  gtts_object.save("E-audio.mp3")
  return Audio("E-audio.mp3")

def ctts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "en", tld = "ca",
                    slow = False)

  gtts_object.save("E-audio.mp3")
  return Audio("E-audio.mp3")
```

# **Practice**

```python
from gtts import gTTS
from pydub import AudioSegment
from pydub.playback import play
import io


stress = "1st" #@param = ["1st","2nd","3rd"]
str = int(stress[0])
n=str-1
wlist = ["Accident, strawberry, seventy, personal, elephant, February, and salary.",
         "Acceptance, vanilla, examine, translation, gorilla, December, and employer.",
         "Accidental, absolute, seventeen, personnel, kangaroo, gasoline, and employee"]


# Your input text
txt = wlist[n]

# Split the text into words
words = txt.split(', ')

# Create an empty audio segment for combining everything
combined = AudioSegment.empty()

# Duration of silence in milliseconds (0.5 seconds)
silence_duration = 500

# Generate audio for each word and add silence
for word in words:
    # Use gTTS to synthesize audio from the text
    tts = gTTS(text=word, lang='en')
    # Save the synthesized audio to a buffer
    mp3_fp = io.BytesIO()
    tts.write_to_fp(mp3_fp)
    mp3_fp.seek(0)
    # Load this audio as an AudioSegment
    word_audio = AudioSegment.from_file(mp3_fp, format="mp3")
    # Add the audio segment to the combined audio
    combined += word_audio
    # Add silence after each word, except the last one
    if word != words[-1]:
        combined += AudioSegment.silent(duration=silence_duration)

# Export the combined audio to a file
combined.export("/content/combined_audio.mp3", format="mp3")

# If you want to play the audio directly in Colab, uncomment the following line:
Audio("/content/combined_audio.mp3")



```
